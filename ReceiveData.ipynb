```python
def getResponse():
    response = requests.get("https://api.thingspeak.com/channels/662426/feeds.json?api_key=59BY6Y6401J12B1P&results=3")
    #print(response.status_code)     #if 200 then working fine
    dataByte = response.content
    d = json.loads(dataByte.decode('utf-8'))
    print("Data Type is here : ",type(d))
    #print(d)
    return(d)
```

```python
def getData():
    ldr1,ldr2,ldr3 = None,None,None
    while(ldr1==None and ldr2==None and ldr3==None):
        data = getResponse()
        l = [data['feeds'][0]['field1'],data['feeds'][0]['field2'],data['feeds'][0]['field3']]
        print(l)
        index = -1
        if('0' in l):
            index = l.index('0')
            print("\nAll is working Fine\n")
        elif('1023' in l):
            index = l.index('1023')
            print("\n\n\n",'-'*50)
            print("\nFault at Pole No. {}".format(l.index('1023')+1))
            print("\n\n\n",'-'*50)
        if(index==0):
            ldr1 = l[index]
        elif(index==1):
            ldr2 = l[index]
        elif(index==2):
            ldr3 = l[index]
    return(ldr1,ldr2,ldr3)
    
```

```python
#Excel Sheet Handling
if __name__ == "__main__": 
    workbook = xlsxwriter.Workbook('LDR_Data.xlsx') 
    worksheet = workbook.add_worksheet("My sheet") 
    worksheet.write('A1','LDR 1')
    worksheet.write('B1','LDR 2')
    worksheet.write('C1','LDR 3')
    worksheet.write('D1','LDR 4')
    worksheet.write('E1','LDR 5')
    row,col = 1,0
    for i in range(30):
        time.sleep(10)
        data = [getData()]
        for l1,l2,l3 in data:
            worksheet.write(row, col, l1)
            worksheet.write(row, col+1, l2)
            worksheet.write(row, col+2, l3)
            row+=1
    workbook.close()
```
